```python
import numpy as np
```

```python
f1 = open("./data/0.txt")
lines = f1.readlines()
len(lines)
```

```python
lines[0]
```

```python
f2 = open("./data/output1.txt", 'w')
```

```python
for line in lines:
    output = []
    fields = line.split("\t")
    for i in range(len(fields)):
        if(i == 3):
            fields[i] = fields[i].replace(" ","")
        if(i < 9):
            if(i == len(fields) - 1):
                output.append(fields[i])
            else:
                output.append(fields[i])
                output.append("\t")
        else:
            if(i == len(fields) - 1):
                output.append(fields[i])
            else:
                output.append(fields[i])
                output.append("&")
    print(output)
    str1 = ""
    for i in range(len(output)):
        str1 += output[i]
    f2.write(str1)
```

```python
f1.close()
f2.close()
```
